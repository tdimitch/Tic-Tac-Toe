```python
# Two player Tic Tac Toe game in Python.
import random

''' We will make the board using dictionary in which 
    keys ('1' to '9') will be the location(i.e : top-left,mid-right,etc.).
    theBoard is initialized with empty spaces and updated with players' moves. '''

theBoard = [' ' for i in range(9)]


    
''' We will have to print the updated board after every move in the game and 
    thus we will make a function in which we'll define the printBoard function
    so that we can easily print the board everytime by calling this function. '''
 
def printBoard(board):
    print('   |   |')
    print(' ' + board[0] + ' | ' + board[1] + ' | ' + board[2])
    print('   |   |')
    print('-----------')
    print('   |   |')
    print(' ' + board[3] + ' | ' + board[4] + ' | ' + board[5])
    print('   |   |')
    print('-----------')
    print('   |   |')
    print(' ' + board[6] + ' | ' + board[7] + ' | ' + board[8])
    print('   |   |')



''' A function to check if a player has won. '''

def checkWin(board):
    # We define the boolean conditions to win
    a1 = board[0] == board[1] == board[2] != ' ' # top row
    a2 = board[3] == board[4] == board[5] != ' ' # middle row
    a3 = board[6] == board[7] == board[8] != ' ' # bottom row
    a4 = board[0] == board[3] == board[6] != ' ' # left column
    a5 = board[1] == board[4] == board[7] != ' ' # middle column
    a6 = board[2] == board[5] == board[8] != ' ' # right column
    a7 = board[0] == board[4] == board[8] != ' ' # diagonal from up-left
    a8 = board[6] == board[4] == board[2] != ' ' # diagonal from down-left
    
    # If any of the winning conditions is verified, return True
    if a1 or a2 or a3 or a4 or a5 or a6 or a7 or a8:
        return True

    
    
''' Create computer moves. '''
def get_pc_move(theBoard):
    
    a = list(range(1,10))
    random.shuffle(a)
    
    for i in range(9):
        if theBoard[a[i] - 1] == ' ':
            return a[i]

        
    
''' The main function game has all the gameplay functionality. '''

def game():

    turn = 'X'
    move_counter = 0
    
    # Ask player if wants to play against computer
    print("Do you want to play against PC? (y/n)")
    play_pc = input()
    
    # If so, who starts?
    if play_pc == 'y' or play_pc == 'Y':
        print("Player X starts, do you want to play first? (y/n)")
        first_turn = input()
        
        if first_turn == 'y' or first_turn == 'Y':
            turn = 'X'
        
        else:
            turn = 'O'
    
    while move_counter <= 9:
        
        printBoard(theBoard)
        print("\nIt's your turn, player " + turn + ". Choose a move: ")

        
        # If computer's turn, call get_pc_move
        if (play_pc == 'y' or play_pc == 'Y') and turn == 'O':
            move = get_pc_move(theBoard)
        
        else:
            move = int(input())

        # Count move or ask for a different move
        if theBoard[move - 1] == ' ':
            theBoard[move - 1] = turn
            move_counter += 1
        else:
            print("That place is already filled. Choose a valid move. ")
            continue
        
        # If the number of moves is >= 5, check if the player has won.
        if move_counter >= 5 and checkWin(theBoard) == True:
            printBoard(theBoard)
            print("\nGame Over.\n **** Player " + turn + " won! ****")
            break

        # If neither X nor O wins and the board is full, we'll declare the result as 'tie'.
        if move_counter >= 9 :
            printBoard(theBoard)
            print("\nGame Over. It's a Tie!")
            break
        
        # Now we have to change the player after every move.
        if turn =='X':
            turn = 'O'
        else:
            turn = 'X'        
    
    # Now we will ask if player wants to restart the game or not. If so, we initialize theBoard.
    restart = input("Do want to play again? (y/n): ")
    if restart == "y" or restart == "Y":  
        for i in range(9):
            theBoard[i] = " "
        print("\n\n")
        game()



game()
```
